```python
import pymongo
import os
import csv
import pandas as pd
from dotenv import load_dotenv
from tqdm import tqdm
import datetime
```

```python
load_dotenv()
mongo_url = os.environ.get("mongo_url")
myclient = pymongo.MongoClient(mongo_url)
mydb = myclient["twitter"]
mycol = mydb["sentiments_state"]
```

```python
df[['state','city', 'sadness', 'joy', 'fear', 'disgust', 'anger']].groupby(['state','city'], as_index=False).mean().to_csv("./sentiments_city.csv")
```

```python
df[['state','city', 'sadness', 'joy', 'fear', 'disgust', 'anger']].groupby(['state','city'], as_index=False).mean()
```

```python
for i, row in tqdm(df[['state', 'city', 'sadness', 'joy', 'fear', 'disgust', 'anger']].groupby(['state', 'city'], as_index=False).mean().iterrows()):
    state = row['state']
    city = row['city']
    sadness = float(row['sadness'])
    joy = float(row['joy'])
    fear = float(row['fear'])
    disgust = float(row['disgust'])
    anger = float(row['anger'])

    result = {'state': state, 'city': city, 'sadness': sadness, 'joy': joy,
              'fear': fear, 'disgust': disgust, 'anger': anger}
    mycol.insert_one(result)
```

```python
for i, row in tqdm(df[['date', 'time', 'tweet', 'link', 'state', 'city', 'sadness', 'joy', 'fear', 'disgust', 'anger']].iterrows()):
    date = row['date']
    time = row['time']
    tweet = row['tweet']
    link = row['link']
    state = row['state']
    city = row['city']
    sadness = float(row['sadness'])
    joy = float(row['joy'])
    fear = float(row['fear'])
    disgust = float(row['disgust'])
    anger = float(row['anger'])

    date_time = date + ' ' + time
    date_time = datetime.datetime.strptime(date_time, '%Y-%m-%d %H:%M:%S')

    result = {'date_time': date_time, 'tweet': tweet, 'link': link, 'state': state, 'city': city,
              'sadness': sadness, 'joy': joy, 'fear': fear, 'disgust': disgust, 'anger': anger}
    
    mycol.insert_one(result)
```
