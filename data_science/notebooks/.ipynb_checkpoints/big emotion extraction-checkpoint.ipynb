```python
import numpy as np
import pandas as pd
import csv
import os
import json
import re
from tqdm import tqdm
from ibm_watson import NaturalLanguageUnderstandingV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
from ibm_watson.natural_language_understanding_v1 import Features, EmotionOptions
```

```python
from dotenv import load_dotenv
load_dotenv()

ibm_api = os.environ.get("ibm_api")
ibm_url = os.environ.get("ibm_url")
```

```python
df = pd.read_csv("../datasets/automation/stage1/combined.csv")
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.columns
```

```python
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'[^\w]', ' ', x))
```

```python
df[['date', 'time', 'tweet', 'link', 'state', 'city']]
```

```python
authenticator = IAMAuthenticator(ibm_api)
natural_language_understanding = NaturalLanguageUnderstandingV1(
    version='2019-07-12',
    authenticator=authenticator
)

natural_language_understanding.set_service_url(ibm_url)
```

```python
def analyze(tweet, target):
    response = natural_language_understanding.analyze(text=tweet, features=Features(
        emotion=EmotionOptions(targets=[target]))).get_result()
    return response['emotion']['document']['emotion']
```

```python
analyze('i love mango', target='mango')
```

```python
output = pd.DataFrame()

for index, row in tqdm(df[['date', 'time', 'tweet', 'link', 'state', 'city']].iterrows()):

    if(index % 20 == 0):
        print("At Index: {}".format(index))

    date = row['date']
    time = row['time']
    tweet = row['tweet']
    link = row['link']
    state = row['state']
    city = row['city']

    try:
        result = analyze(tweet, target=city)
        result['date'] = date
        result['time'] = time
        result['tweet'] = tweet
        result['link'] = link
        result['state'] = state
        result['city'] = city
        output = output.append(result, ignore_index=True)
    except:
        result = {'sadness': np.nan, 'joy': np.nan, 'fear': np.nan, 'disgust': np.nan, 'anger': np.nan}
        result['date'] = date
        result['time'] = time
        result['tweet'] = tweet
        result['link'] = link
        result['state'] = state
        result['city'] = city
        output = output.append(result, ignore_index=True)
```
