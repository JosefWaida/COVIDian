```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("../datasets/automation/india_info.csv")
```

```python
india_info = df[['state', 'city']]
```

```python
india_info
```

```python
output_file = "{}/{}.csv".format("state", "city")
```

```python
#!twint -s "Unlock1" --limit 20 -o output_file --csv 
```

```python
bash_command_list = []

for i, row in india_info.iterrows():
    state, city = row['state'], row['city']
    temp = "twint -s 'Unlock1 AND {}' --limit 400 -o '{}--{}.csv' --csv".format(city, state, city)
    bash_command_list.append(temp)
```

```python
bash_command_list
```

```python
len(bash_command_list)
```

```python
file1 = open("bash_command_list.txt","w")
file1.writelines(bash_command_list) 
file1.close()
```

```python
with open('bash_command_list.txt', 'w') as f:
    for command in bash_command_list:
        f.write("%s\n" % command)
```

```python
#df.sort_values(by=['col1'])
india_info.sort_values(by=['state'])
```

```python
scrapped_data = os.listdir("./automated_dataset/")
scrapped_data.remove(".DS_Store")
scrapped_data.remove("bash_command_list.txt")
scrapped_data.remove("stage1")
```

```python
scrapped_data
```

```python
output_df = pd.DataFrame()

for file in scrapped_data:
    file_df = pd.read_csv('./automated_dataset/{}'.format(file))
    
    temp = file.split(".")[0].split("--")
    state, city = temp[0], temp[1]
    
    file_df['state'] = state
    file_df['city'] = city
    
    output_df = output_df.append(file_df)
```

```python
output_df.to_csv('./automated_dataset/stage1/combined.csv')
```
